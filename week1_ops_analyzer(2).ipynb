# Week 1 Operations Analyzer
This notebook demonstrates operational efficiency tracking and explains the logic for a junior analyst.

## Step 1: KPI Function
We calculate efficiency by comparing actual output against target output.

```
def calculate_efficiency(actual_output, target_output):
    return (actual_output / target_output) * 100
```

## Step 2: Status Classifier
We categorize efficiency levels into Critical, Warning, and Normal.

```
def get_operational_status(efficiency):
    if efficiency < 70:
        return "Critical"
    elif 70 <= efficiency <= 89:
        return "Warning"
    else:
        return "Normal"
```

## Step 3: Data Modeling
This example uses three retail stores as operational sites.

```
sites = [
    {"name":"Retail Store A","actual_output":850,"target_output":1000},
    {"name":"Retail Store B","actual_output":620,"target_output":1000},
    {"name":"Retail Store C","actual_output":940,"target_output":1000}
]
```

## Step 4: Execution
We iterate through the sites, calculate efficiency, determine status, and print results.

```
for site in sites:
    efficiency = calculate_efficiency(
        site["actual_output"],
        site["target_output"]
    )

    status = get_operational_status(efficiency)

    print("Operational Site Report")
    print("-" * 30)
    print(f"Site: {site['name']}")
    print(f"Actual Output: {site['actual_output']}")
    print(f"Target Output: {site['target_output']}")
    print(f"Efficiency: {efficiency:.2f}%")
    print(f"Status: {status}")
    print("-" * 30)
```
